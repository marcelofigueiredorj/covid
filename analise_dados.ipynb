```python
df = pd.read_csv('dados_covid.csv', sep=';')
```

```python
dados_brasil = df[df['estado'].isna()]
```

```python
fig = px.bar(dados_brasil, x='data', y='obitosAcumulado')
fig.show()
```

```python
dados_por_semana = dados_brasil.groupby('semanaEpi').sum()
fig_dados_por_semana = px.bar(dados_por_semana, y='casosAcumulado')
fig_dados_por_semana.show()
```

## Dados por Estado

```python
filtro_estado = 'SP'
dados_estado = df[df['estado'].notnull() & df['municipio'].isna() & (df['estado'] == filtro_estado)]
dados_estado
```

```python
estados = df['estado'].unique()
estados
```

```python
fig_estado = px.bar(dados_estado, x='data', y='obitosAcumulado')
fig.show()
```

## Dados por Região

```python
regioes = df['regiao'].unique()
regioes
```

```python
filtro_regiao = 'Nordeste'
dados_regiao = df[df['municipio'].isna() & (df['regiao'] == filtro_regiao)]
dados_regiao = dados_regiao.groupby('estado').sum().sort_values('casosAcumulado', ascending=False)
#dados_regiao['estado'].unique()
dados_regiao
```

```python
fig_regiao = px.bar(dados_regiao, x=dados_regiao.index, y='casosAcumulado')
fig_regiao.show()
```

```python
dados_geral_brasil = df[df['municipio'].isna()]
dados_geral_brasil = dados_geral_brasil.groupby('regiao').sum().sort_values('casosAcumulado', ascending=False)
dados_geral_brasil.drop('Brasil', inplace=True)
dados_geral_brasil
```

```python
fig_geral_brasil = px.bar(dados_geral_brasil, x=dados_geral_brasil.index, y='casosAcumulado')
fig_geral_brasil.show()
```
